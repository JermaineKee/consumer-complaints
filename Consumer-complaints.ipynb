```python
dec_comp[['Complaint ID', 'Date received']]
```

```python
date_series = dec_comp.copy()
```

```python
date_index = date_series.pop('Date received')
date_index.head()
```

```python
date_series.index = pd.to_datetime(date_index, format='%m/%d/%Y')
```

```python
date_series.head()
```

```python
comp_by_day = date_series.copy()
```

```python
comp_by_day.index = comp_by_day.index.dayofweek
```

```python
comp_by_day.index.value_counts()
```

```python
comp_by_day.index.value_counts().plot(title='complaints by day',kind='bar')
plt.show()
```

This graph shows that most complaints were reported at the beginning of the week.  Monday, Tuesday, and Wedensday.
